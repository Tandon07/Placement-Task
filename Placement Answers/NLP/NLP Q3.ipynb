```python
import pandas as pd
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from gensim.models import LdaModel
from gensim.corpora import Dictionary

# Read the CSV file
df = pd.read_csv(r'C:\Users\tando\Downloads\words.csv')

```

```python
# Preprocess the text
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python

def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()
    # Remove special characters and numbers
    text = re.sub(r'[^a-zA-Z]', ' ', text)
    # Tokenize the text
    tokens = word_tokenize(text)
    # Remove stopwords and lemmatize
    tokens = [lemmatizer.lemmatize(token) for token in tokens if token not in stop_words]
    # Join the tokens back to form a string
    preprocessed_text = ' '.join(tokens)
    return preprocessed_text
```

```python

# Apply preprocessing to the 'text' column in the dataframe
df['preprocessed_text'] = df['word'].apply(preprocess_text)
```

```python
import pandas as pd
import re
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

# Read the CSV file
df = pd.read_csv(r'C:\Users\tando\Downloads\words.csv')


```

```python
df=df.dropna()
```

```python
# Keyword Extraction using CountVectorizer
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['word'])

```

```python
# Get the feature names (keywords)
feature_names = vectorizer.get_feature_names()
```

```python

# Print the keywords
print("Keywords:")
print(feature_names)
```

```python
# Topic Modeling using Latent Dirichlet Allocation (LDA)
num_topics = 3  # Number of topics to extract
lda_model = LatentDirichletAllocation(n_components=num_topics, random_state=42)
lda_model.fit(X)
```

```python
# Get the main topics and their keywords
print("\nTopics:")
for idx, topic in enumerate(lda_model.components_):
    topic_keywords = [feature_names[i] for i in topic.argsort()[:-6:-1]]
    print(f"Topic {idx+1}: {topic_keywords}")
```
