```python
import scipy.stats as stats

# Group A data
mean_A = 2.5
std_A = 0.8
n_A = 30

# Group B data
mean_B = 2.2
std_B = 0.6
n_B = 30


# significance level
alpha = 0.05
```

```python
# t-value and p-value
t_stat, p_value = stats.ttest_ind_from_stats(mean_A, std_A, n_A, mean_B, std_B, n_B, equal_var=False)
```

```python
if p_value < alpha:
    print("Reject the null hypothesis.")
    print("There is a significant difference in the mean improvement scores between Group A and Group B.")
else:
    print("Fail to reject the null hypothesis.")
    print("There is no significant difference in the mean improvement scores between Group A and Group B.")
```
