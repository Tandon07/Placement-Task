```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats

# Load the data into a DataFrame


# Calculate the difference in blood pressure
df["Difference"] = df[" Blood Pressure Before (mmHg)"] - df[" Blood Pressure After (mmHg)"]

# Plot a histogram of the differences
plt.hist(df["Difference"], bins=10)
plt.xlabel("Difference in Blood Pressure")
plt.ylabel("Frequency")
plt.title("Distribution of Blood Pressure Differences")
plt.show()

# Perform Shapiro-Wilk test for normality
stat, p_value = stats.shapiro(df["Difference"])
alpha = 0.05  # Significance level

if p_value > alpha:
    print("The change in blood pressure follows a normal distribution.")
else:
    print("The change in blood pressure does not follow a normal distribution.")

# Create a Q-Q plot
stats.probplot(df["Difference"], dist="norm", plot=plt)
plt.xlabel("Theoretical Quantiles")
plt.ylabel("Sample Quantiles")
plt.title("Q-Q Plot of Blood Pressure Differences")
plt.show()
```
