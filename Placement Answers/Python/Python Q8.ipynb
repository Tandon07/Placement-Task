a. Get all the overall ratings for each season and using plots compare the ratings for all the
seasons, like season 1 ratings, season 2, and so on.

```python
# Group the data by season and calculate the average ratings
df3['rating'] = df3['rating'].apply(lambda x: x['average'] if isinstance(x, dict) else None)

# Drop rows with missing ratings
df3 = df3.dropna(subset=['rating'])

# Convert rating column to numeric
df3['rating'] = pd.to_numeric(df3['rating'])
```

```python
df3.head()
```

```python
season_ratings = df3.groupby(['name', 'season'])['rating'].mean()
```

```python
season_ratings
```

```python
# Plot the ratings for each season
import matplotlib.pyplot as plt
season_ratings.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Season')
plt.ylabel('Average Rating')
plt.title('Average Ratings for Each Season')
plt.show()
```

b. Get all the episode names, whose average rating is more than 8 for every season

```python
df=df3.copy()
```

```python
filtered_episodes = df[df['rating'] > 8].groupby(['season'])['name'].apply(list)

# Reset index to convert the grouped data into a DataFrame
filtered_episodes = filtered_episodes.reset_index()
```

```python
filtered_episodes
```

c. Get all the episode names that aired before May 2019

```python
df['airstamp'] = pd.to_datetime(df['airstamp'])

# convert the Timestamp object to UTC timezone
target_date = pd.Timestamp('2019-05-01', tz='UTC')

# filter episodes that aired before May 2019
filtered_episodes = df[df['airstamp'] < target_date]

# get the episode names
episode_names = filtered_episodes['name']
```

```python
# print the episode names
for episode_name in episode_names:
    print(episode_name)
```

d. Get the episode name from each season with the highest and lowest rating

```python
grouped_episodes = df.groupby('season').apply(lambda x: x.loc[x['rating'].idxmax()]['name'])
grouped_episodes_lowest = df.groupby('season').apply(lambda x: x.loc[x['rating'].idxmin()]['name'])
```

```python
# print the episode names with highest and lowest ratings for each season
for season, episode_name in grouped_episodes.items():
    print(f"Season {season}: Highest Rated Episode - {episode_name}")

for season, episode_name in grouped_episodes_lowest.items():
    print(f"Season {season}: Lowest Rated Episode - {episode_name}")
```

e. Get the summary for the most popular ( ratings ) episode in every season

```python
df_sorted = df.sort_values(by='rating', ascending=False)
```

```python
# group the data by season and retrieve the summary of the episode with the highest rating
grouped_episodes = df_sorted.groupby('season').first()['summary']
```

```python
# print the summary for the most popular episode in each season
for season, summary in grouped_episodes.items():
    print(f"Season {season}: Most Popular Episode Summary - {summary}")
```
