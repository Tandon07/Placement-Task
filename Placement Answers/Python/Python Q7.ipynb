```python
import requests
import pandas as pd

url = 'https://data.nasa.gov/resource/y77d-th95.json'
response = requests.get(url)               #downloading the raw json using requests
data = response.json()

```

```python
df = pd.DataFrame(data)    # Convert JSON data to DataFrame
```

```python
output_file_q4 = 'nasa_data.csv'           #creating output file
df.to_csv(output_file_q4, index=False)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
# we can see the majority of :@computed_region_cbhk_fwbd and @computed_region_nnqa_25f4 are missing so we will drop it
```

```python
df=df.drop([':@computed_region_cbhk_fwbd',':@computed_region_nnqa_25f4'],axis=1)
```

```python
# also dropping the single null in year
```

```python
df['fall'].fillna(df['fall'].mode(),inplace=True)
```

```python
df.info()
```

```python
df['mass']=pd.to_numeric(df['mass'])
```

```python
# Split the datetime string and extract the year
df['year'] = df['year'].str.split('-').str[0]
```

```python
# filling mass with average
```

```python
df['mass'].fillna(df['mass'].mean(),inplace=True)
```
