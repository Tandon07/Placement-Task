```python
import pandas as pd

def convert_to_seconds(time_str):
    minutes, seconds = map(int, time_str.split(':'))
    total_seconds = minutes * 60 + seconds
    return total_seconds

df1['spawn_time'] = df1['spawn_time'].apply(convert_to_seconds)
```

```python
for i in range(0,df1.shape[0]):
    
    df1['spawn_time'][i]=convert_to_seconds(df1['spawn_time'][i])
#     print(df1['spawn_time'][i])
```
